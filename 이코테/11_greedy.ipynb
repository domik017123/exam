```python
##모험가 길드 3 5 1 1 3 2 1
n = int(input())
data = list(map(int, input().split()))

group = []
group_count = 0
member_count = 0

#최강자부터 정렬 1 1 1 2 3 3 5
data.sort()
#순서대로 그룹 가입
for fear in data:
  group.append(fear)
  member_count += 1
  #그룹 마감
  if member_count == fear:
    group_count += 1
    member_count = 0


print(group_count)
```

```python
##곱하기 혹은 더하기
data = input()
data_int = [int(char) for char in data]
print(data_int)

result = data_int[0]

for i in range(1, len(data_int)):
  if data_int[i] != 0 and result !=0 :
    result *= data_int[i]
  else:
    result += data_int[i]


print(result)
```

```python
#문자열 뒤집기
data = input()
data_int = [int(char) for char in data]

bound_cnt = 0

for i in range(0, len(data_int)-1):
  if data_int[i] != data_int[i+1]:
    bound_cnt += 1

# 0 b0 0, 01 b1 1, 010 b2 1, 0101 b3 2, 01010 b4 2
if bound_cnt == 0:
  print(0)
else:
  result = (bound_cnt-1)//2 + 1
  print(result)
```

```python
##만들 수 없는 금핵
from itertools import combinations

n = int(input()) # 5
data = list(map(int, input().split())) # 1 1 3 6 13 

data.sort()

# 1 없음
if data[0] != 1:
  print(1)


else:
  sum_list = []
  for c in range(1, n + 1):
    for cb in combinations(data, c):
      combination_sum = sum(cb)
      sum_list.append(combination_sum)
  sum_list.sort()
  print(sum_list)

  max_sum = sum(data)
  min_sum = 1
  for num in range(1,max_sum+2):
    if num not in sum_list:
      min_num = num
      break

  print(min_num)
```

```python
##볼링공 고르기
n, m = map(int,input().split())
data = list(map(int, input().split())) # 

w_list = [0]*m
# print(w_list)
for w in data:
  w_list[w-1] += 1
# print(w_list)

count = 0
for i in range(0,m):
  count += (sum(w_list)-w_list[i])*w_list[i]

print(int(count/2))
```

```python
##무지의 먹방라이브
food_times = [3,1,2]
k = 5
food_num = len(food_times) #음식 가짓수
print('0','0',food_times)


n = 0 #n번째음식
time = 1
while True:
  if time == k+1:
    break

  elif food_times[n%food_num] > 1: #남은 음식
    food_times[n%food_num] -=1 #먹기
    print(time,n%food_num+1,food_times)
    
    n+=1  #다음 음식
    time+=1 #시간 흐름

  elif food_times[n%food_num] == 1:#한개 남을때
    food_times[n%food_num] =-1 #다먹음
    print(time,n%food_num+1,food_times)

    n+=1  #다음 음식
    time+=1 #시간 흐름

  else: #없을때
    food_times[n%food_num] =-1 #다먹음
    n +=1

print(n%food_num+1)
```

# 새 섹션

```python
def mukbang(food_times, k):
    food_num = len(food_times) #음식 가짓수
    
    n = 0 #n번째음식
    time = 1
    while True:
        if time == k+1:
            break

        elif food_times[n%food_num] > 1: #남은 음식
            food_times[n%food_num] -=1 #먹기

            n+=1  #다음 음식
            time+=1 #시간 흐름

        elif food_times[n%food_num] == 1:#한개 남을때
            food_times[n%food_num] =-1 #다먹음

            n+=1  #다음 음식
            time+=1 #시간 흐름

        else: #없을때
            food_times[n%food_num] =-1 #다먹음
            n +=1

    answer = n%food_num+1
    return answer
```
