```python
#문자열 압축

def solution(s):
  answer = 0

  # 비교할 블록 정의
  def str_block(start_idx, len_block, s):
    return s[start_idx:start_idx+len_block]

  # 압축 문자열 비교 리스트
  comp_list = []
  for i in range(1, len(s)): #len_block = i
    # 블록 인덱싱 초기화
    start_idx = 0
    next_idx = start_idx + i
    block_cnt = 1

    # 압축 문자열 초기화    
    comp = ''
    while start_idx < len(s):    
      # 블록 동일 
      if str_block(start_idx, i, s) == str_block(next_idx, i, s):
        block_cnt += 1
        start_idx = next_idx
        next_idx += i

      else:
        # 이전 블록이 압축된 경우
        if block_cnt>1:
          comp += str(block_cnt) + str_block(start_idx, i, s)
          block_cnt = 1

          start_idx = next_idx##
          next_idx = start_idx+i

        # 이전 블록이 압축되지 않은 경우 
        else:
          comp += s[start_idx]
          start_idx += 1##
          next_idx = start_idx+i

    comp_list.append(len(comp))

  answer = min(comp_list)
  return answer


s = 'xababcdcdababcdcd'
result = solution(s)
print(result)
```

```python
##기둥과 보 설치
import numpy as np
##def check(): return T/F
def check(x,y,a,build):
    #case1.기둥 x y 0 1 >> b[x,y]=0
    if a==0:
        #바닥 위/보의 끝/기둥 위
        if y==0 or build[x-1][y]==1 or build[x][y]==1 or build[x][y-1]==0:
            return True

    #case2.보 x y 1 1 >> b[x,y]=1
    if a==1:
        #기둥 위/양쪽 보
        if build[x][y-1]==0 or build[x+1][y-1]==0 or (build[x-1][y]==1 and build[x+1][y]==1):
            return True

##def delete(): return T/F
def delete(x,y,a,build):
    build[x][y] = -1
#기둥 삭제 #x,y >check([x,y+1]:case1,2 , [x-1,y+1]:case2)
    if check(x,y+1,0,build) and check(x,y+1,1,build) and check(x-1,y+1,1,build):
        return True

#보 삭제 #x,y >check([x-1,y]:case2 , [x,y]:case1, [x+1,y]:case1,2)
    if check(x-1,y,1,build) and check(x,y,0,build) and check(x+1,y,0,build) and check(x+1,y,1,build):
        return True

##def solution(): return result<<[x,y,b[x,y]]
def solution(n, build_frame):
    build = np.array([[-1]*(n+1) for _ in range(n+1)])
    print(build)
    #bulid_frame 입력
    for i in range(len(build_frame)):
        x,y,a,b = build_frame[i]

        # check()==T 
        if b==1:
            if check(x,y,a,build)==True:
                build[x][y]=a
                print("build",x,y,a)
            else:
                print("denied",x,y,a)

        # delete()==T
        if b==0:
            if delete(x,y,a,build)==True:
                # build[x][y]=-1
                print("del",x,y,a)
            else:
                build[x][y]=a
                print("denied",x,y,a)
                
        print(build)

    #result 출력        
    result = [[x,y,build[x][y]] for x in range(n+1) for y in range(n+1) \
              if build[x][y]!=-1]              
    return result

solution(5,[[0, 0, 0, 1], [2, 0, 0, 1], [4, 0, 0, 1], [0, 1, 1, 1], [1, 1, 1, 1], [2, 1, 1, 1], [3, 1, 1, 1], [2, 0, 0, 0], [1, 1, 1, 0], [2, 2, 0, 1]])
```

```python
##치킨배달
from itertools import combinations
#지도 생성
# n, m = map(int, input().split())
# ch_map = [[0]*n for _ in range(n)]
# for i in range(n):
#     ch_map[i] = list(map(int, input().split()))

n = 5
m = 2
ch_map = [[0, 2, 0, 1, 0], [1, 0, 1, 0, 0], [0, 0, 0, 0, 0], [2, 0, 0, 1, 1], [2, 2, 0, 1, 2]]


#치킨집 개수 중 M개 선택
ch_num=0
ch_list=[]
for i in range(n):
    for j in range(n):
        if ch_map[i][j]==2:
            ch_num +=1
            ch_list.append([i,j])
print(ch_num,ch_list)

live_ch_list=list(combinations(ch_list,m))
print(len(live_ch_list))

#치킨거리 계산 >>>가장 가까운 치킨집 구현 안함
def get_dist(live_stores, ch_map):
    for k in range(m):
        x, y = live_stores[k][0], live_stores[k][1]
        ch_dist = 0
        for i in range(n):
            for j in range(n):
                if ch_map[i][j]==1:
                    a, b = i, j
                    ch_dist += abs(a-x)+abs(b-y)
    return ch_dist

ch_dist_list=[]
for _ in range(len(live_ch_list)):
    ch_dist = get_dist(live_ch_list[_], ch_map)
    ch_dist_list.append(ch_dist)
print(len(ch_dist_list),ch_dist_list)
#최소
result = min(ch_dist_list)
print(result)
```
