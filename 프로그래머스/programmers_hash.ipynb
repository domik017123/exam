```python
#1.폰켓몬 lv1
'''
nums는 폰켓몬의 종류 번호가 담긴 1차원 배열입니다.
nums의 길이(N)는 1 이상 10,000 이하의 자연수이며, 항상 짝수로 주어집니다.
폰켓몬의 종류 번호는 1 이상 200,000 이하의 자연수로 나타냅니다.
가장 많은 종류의 폰켓몬을 선택하는 방법이 여러 가지인 경우에도,
선택할 수 있는 폰켓몬 종류 개수의 최댓값 하나만 return 하면 됩니다.
'''

def solution(nums):
    answer = 0

    # m=N/2
    m = len(nums)//2
    #nums_set = set 자료형으로 변환
    nums_set = set(nums)
    #p_max = set 자료형 길이 : 폰켓몬 가짓수
    p_max = len(nums_set)
    #1. n/2마리가 p_max보다 작거나 같을 때 : answer = n/2
    if m <= p_max:
        answer = m
    #2. 클때 : answer = p_max
    else:
        answer = p_max

    return answer

'''
23.07.30. 15min pass
'''
```

```python
#2.완주하지 못한 선수 lv1
'''
마라톤 경기에 참여한 선수의 수는 1명 이상 100,000명 이하입니다.
completion의 길이는 participant의 길이보다 1 작습니다.
참가자의 이름은 1개 이상 20개 이하의 알파벳 소문자로 이루어져 있습니다.
참가자 중에는 동명이인이 있을 수 있습니다.
'''


def solution(participant, completion):
    answer = ''

    #정렬후 비교할때 다른 경우의 participant 리턴
    participant.sort()
    completion.sort()
    completion.append("NaN")

    n = len(participant)
    for i in range(n):
        if participant[i] != completion[i]:
            answer = participant[i]
            break

    return answer

'''
230730 16min pass

5min > 시간초과

#참가자에서 완주자 요소 하나씩 빼기 : for문(N)*remove(N) > O(N^2)
for i in completion:
    participant.remove(i)

+11min > 통과
#정렬후 비교할때 다른 경우의 participant 리턴 : 정렬(NlogN) + for문(N) * !=(1) > O(NlogN))

'''
```

```python
#3.전화번호 목록 lv2
'''
phone_book의 길이는 1 이상 1,000,000 이하입니다.
각 전화번호의 길이는 1 이상 20 이하입니다.
같은 전화번호가 중복해서 들어있지 않습니다.
'''
from collections import deque

def solution(phone_book):
    answer = True

    #정렬 후 작은 수 부터 확인
    phone_book.sort()    ## NlogN
    ph = deque(phone_book)

    while ph:            ##for문 while문 중첩시 NlogN
        x = ph.popleft()

    #작은 수 길이로 슬라이싱
        for y in ph:
            if x == y[:len(x)]:
                answer = False

            return answer

    return answer

'''
230730 18min fail

채점 결과
정확성: 70.8
효율성: 12.5
합계: 83.3 / 100.0
'''
```

```python
#3.전화번호 목록 lv2
def solution(phone_book):
    answer = True

    #정렬 후 작은 수 부터 확인
    phone_book.sort()    ## NlogN

    #작은 수 길이로 슬라이싱
    n = len(phone_book)
    # i : 0 ~ n-2
    for i in range(n-1):    ## N^2
        x = phone_book[i]
        # j : i+1 ~ n-1
        for j in range(i+1,n):
            if x == phone_book[j][:len(x)]:
                answer = False

                return answer

    return answer
'''
230731 15min fail
채점 결과
정확성: 83.3
효율성: 8.3
합계: 91.7 / 100.0
'''
```

```python
#4.의상 lv2
'''
clothes의 각 행은 [의상의 이름, 의상의 종류]로 이루어져 있습니다.
코니가 가진 의상의 수는 1개 이상 30개 이하입니다.
같은 이름을 가진 의상은 존재하지 않습니다.
clothes의 모든 원소는 문자열로 이루어져 있습니다.
모든 문자열의 길이는 1 이상 20 이하인 자연수이고 알파벳 소문자 또는 '_' 로만 이루어져 있습니다.
'''


from collections import Counter

def solution(clothes):
    answer = 1


    # 의상 종류 가짓수 구하기 : Counter 사용
    clothes_category = [x[1] for x in clothes]
    dict_clothes = dict(Counter(clothes_category))

    # ans : 종류별 의상의 (가짓수+1) 모두 곱하기 - 1
    num_category = list(dict_clothes.values())

    for i in num_category:
        answer *= i+1
    answer -= 1

    return answer

'''
230730 20min pass
'''
```

```python
#5. 베스트 앨범 lv3
'''
genres[i]는 고유번호가 i인 노래의 장르입니다.
plays[i]는 고유번호가 i인 노래가 재생된 횟수입니다.
genres와 plays의 길이는 같으며, 이는 1 이상 10,000 이하입니다.
장르 종류는 100개 미만입니다.
장르에 속한 곡이 하나라면, 하나의 곡만 선택합니다.
모든 장르는 재생된 횟수가 다릅니다.

>> NlogN
'''

def solution(genres, plays):
    answer = []

    ## 장르 종류별 재생횟수 구하기 : {'classic': 1450, 'pop': 3100}
    set_g = set(genres)
    dict_g = {}
    #종류
    for g in set_g:
        #재생횟수
        cnt_g = 0
        for i in range(len(genres)):
            if genres[i] == g:
                cnt_g += plays[i]
        dict_g[g] = cnt_g

    ## 데이터 묶기[idx,genres,plays] : [[0, 1450, 500], [1, 3100, 600], [2, 1450, 150], [3, 1450, 800], [4, 3100, 2500]]
    data = []
    for idx in range(len(genres)):
        genres[idx] = dict_g[genres[idx]] #장르 숫자로 변환
        data.append([idx,genres[idx],plays[idx]])

    ## 장르, 재생횟수, 인덱스 순서대로 정렬 : [[4, 3100, 2500], [1, 3100, 600], [3, 1450, 800], [0, 1450, 500], [2, 1450, 150]]
    data = sorted(data, key = lambda x : (-x[1], -x[2], x[0]))

    ## 장르 개수 2이하 일때 인덱스 출력
    cnt = 1
    cnt_g = -1 #확인할 장르
    for i in range(len(data)):
        d = data[i]
        #장르 같은지 확인
        if d[1]==cnt_g:
            cnt+= 1
        else:
            cnt_g = d[1]
            cnt = 1
        #카운트 2이하 확인
        if cnt <= 2:
            answer.append(d[0])
        else:
            continue

    return answer


'''
230802 51min pass
'''
```
